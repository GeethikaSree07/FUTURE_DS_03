```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r'C:/Users/HP/OneDrive/Desktop/Internship/FUTURE_DS_03/data/bank-additional-full.csv', sep=';')
df.head()
conversion_rate = (df['y'].value_counts(normalize=True) * 100)

print(conversion_rate)
sns.countplot(x='y', data=df)

plt.title('Customer Conversion')

plt.savefig('C:/Users/HP/OneDrive/Desktop/Internship/FUTURE_DS_03/images/conversion_rate.png')
plt.show()
```

```python
channel_conversion = pd.crosstab(df['contact'], df['y'])

print(channel_conversion)
channel_conversion.plot(kind='bar')

plt.title('Conversion by Contact Type')
plt.ylabel('Count')


plt.savefig('C:/Users/HP/OneDrive/Desktop/Internship/FUTURE_DS_03/images/channel_conversion.png')
plt.show()
```

```python
campaign_analysis = pd.crosstab(df['campaign'], df['y'])

print(campaign_analysis.head())

bins = [18, 25, 35, 45, 55, 65, 100]

labels = ['18-25', '26-35', '36-45', '46-55', '56-65', '65+']

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

age_conversion = pd.crosstab(df['age_group'], df['y'])

print(age_conversion)

age_conversion.plot(kind='bar')

plt.title('Conversion by Age Group')

plt.tight_layout()

plt.savefig('C:/Users/HP/OneDrive/Desktop/Internship/FUTURE_DS_03/images/age_conversion.png')

plt.show()
```

```python
job_analysis = pd.crosstab(df['job'], df['y'])

print(job_analysis)
job_analysis.plot(kind='bar', figsize=(12,6))

plt.title('Conversion by Job Type')

plt.xlabel('Job')

plt.ylabel('Count')

plt.tight_layout()

plt.savefig('C:/Users/HP/OneDrive/Desktop/Internship/FUTURE_DS_03/images/job_analysis.png')

plt.show()
```

# Key Insights

1. Most customers did not convert successfully during marketing campaigns.

2. Cellular contact method generated more successful conversions than telephone contact.

3. Customers in middle age groups showed better conversion rates.

4. Higher campaign frequency reduced customer response effectiveness.

5. Funnel analysis revealed significant drop-off before customer conversion.

# Recommendations

1. Focus more on cellular-based marketing campaigns.

2. Improve lead nurturing strategies to reduce funnel drop-off.

3. Avoid excessive repeated campaign calls.

4. Target high-converting customer age groups.

5. Optimize marketing funnel stages for better conversion performance.
